```python
# lets create a function to deal with M KB and others 
# def size_to_kb(size):
#     if size.endswith('M'):
#         return float(size[:-1]) * 1024 # convert M to KB
#     elif size.endswith('k'):
#         return float(size[:-1]) # already in KB
#     else:
#         return np.nan # for 'Varies with device' and other non numeric values
def size_process(item):
    if str(item)[-1] == 'M':
        res = float(str(item).replace('M', ''))
        res = res * 1024
        return res
    elif str(item)[-1] =='k':
        res = float(str(item).replace('k', ''))
        return res
    else:
        return np.nan
```

```python
#test 
size_process('19M') # output is : 19456.0
```

```python
# apply the function to the Size column
df_work['Size'] = df_work['Size'].apply(size_process)
df_work['Size'].unique()
```

```python
# lets see the dataframe 
df_work[df_work['App'].duplicated(keep = 'first')] # keep = first 
# keep parameter in .duplicated() and .drop_duplicates()
# -------------------------------------------------------
# keep tells pandas which occurrence of a duplicate to KEEP and which to MARK/DROP
#
# keep = 'first' (DEFAULT)
#   → keeps the FIRST occurrence
#   → marks all subsequent duplicates as True
#   → use this when dropping duplicates (most common)
#
# keep = 'last'
#   → keeps the LAST occurrence
#   → marks all previous duplicates as True
#   → use when the latest entry is more updated/correct
#
# keep = False
#   → marks EVERY occurrence as True (including the first)
#   → keeps NOTHING, drops ALL copies entirely
#   → use only when you want to INSPECT all duplicate rows
#   → never use this for dropping if you want to retain one copy
#
# RULE OF THUMB:
#   inspecting duplicates  → keep = False
#   dropping duplicates    → keep = 'first' (safe default)
```

```python
# drop duplicates based on App column
df_work.drop_duplicates(subset='App', keep='first', inplace=True)

#subset = 'App'
# tells pandas to check duplicates ONLY based on App column
# if two rows have same App name → considered duplicate
# without subset → pandas checks ALL columns must be identical to consider it a duplicate


# Why subset='App' and not just .drop_duplicates()?
#        df_work.drop_duplicates()  
# only drops rows where EVERY column is identical
# two rows with same App but different Rating → NOT considered duplicate

#       df_work.drop_duplicates(subset='App')  
# drops rows where App name is same
# even if other columns differ → still considered duplicate
```
