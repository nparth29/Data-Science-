````python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder

data = {
    'Age':        [25, 34, 28, 45, 31],
    'City':       ['Mumbai', 'Delhi', 'Chennai', 'Mumbai', 'Kolkata'],
    'Gender':     ['Male', 'Female', 'Male', 'Female', 'Male'],
    'Purchased':  ['Yes', 'No', 'Yes', 'Yes', 'No'],
    'Salary':     [35000, 95000, 18000, 75000, 52000]
}
df = pd.DataFrame(data)
df
````

````python
#Now apply One Hot Encoding on City and Gender — both nominal, low unique values:
````

````python
# Step 1 — identify columns to encode
ohe_cols = ['City', 'Gender']

# Step 2 — fit and transform
encoder = OneHotEncoder(sparse_output=False)
encoded_array = encoder.fit_transform(df[ohe_cols])

# Step 3 — convert to DataFrame
encoded_df = pd.DataFrame(encoded_array, columns=encoder.get_feature_names_out())

# Step 4 — drop original columns and join encoded ones
df_encoded = df.drop(columns=ohe_cols)
df_final = pd.concat([df_encoded, encoded_df], axis=1)
df_final
````

````python
""" 
pd.concat([df1, df2], axis=?)
```

`[df1, df2]` — a **list** of DataFrames to join together.

---

### axis=0 vs axis=1 — The Most Important Part

Think of your DataFrame as a table:
```
axis=0 → vertical direction   (along rows    ↕)
axis=1 → horizontal direction (along columns ↔)
```

Visually:
```
axis = 0  (stacking rows on top of each other)

   df1:          df2:          result:
   A   B         A   B         A   B
   1   2    +    5   6    =    1   2
   3   4         7   8         3   4
                               5   6
                               7   8
```
```
axis = 1  (joining columns side by side)

   df1:          df2:          result:
   A   B         C   D         A   B   C   D
   1   2    +    5   6    =    1   2   5   6
   3   4         7   8         3   4   7   8

"""
````

````python
#ordinal : 
data = {
    'Name':          ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'Education':     ['high school', 'graduate', 'phd', 
                      'post graduate', 'graduate'],
    'Experience':    ['junior', 'senior', 'junior', 'lead', 'senior'],
    'Satisfaction':  ['low', 'high', 'medium', 'high', 'low'],
    'Salary':        [35000, 95000, 18000, 75000, 52000]
}
df = pd.DataFrame(data)
df
````

````python
from sklearn.preprocessing import OrdinalEncoder

# Define order for EACH ordinal column separately
education_order   = ["high school", "graduate", "post graduate", "phd"]
experience_order  = ["junior", "senior", "lead"]
satisfaction_order = ["low", "medium", "high"]

ordinal_cols = ["Education", "Experience", "Satisfaction"]

ordinal_encoder = OrdinalEncoder(categories=[
    education_order,     # order for Education column
    experience_order,    # order for Experience column
    satisfaction_order   # order for Satisfaction column
])

encoded_array = ordinal_encoder.fit_transform(df[ordinal_cols])
encoded_df = pd.DataFrame(encoded_array, columns=[
    "Education_encoded", "Experience_encoded", "Satisfaction_encoded"
])

# drop originals and join encoded
df_final = pd.concat([df.drop(columns=ordinal_cols), encoded_df], axis=1)
df_final

# ```

# **Result:**
# ```
#       Name  Salary  Education_encoded  Experience_encoded  Satisfaction_encoded
# 0    Alice   35000                0.0                 0.0                   0.0
# 1      Bob   95000                1.0                 1.0                   2.0
# 2  Charlie   18000                3.0                 0.0                   2.0
# 3    David   75000                2.0                 2.0                   2.0
# 4      Eve   52000                1.0                 1.0                   0.0
# ```

# **Mapping happening internally:**
# ```
# Education:    high school=0  graduate=1  post graduate=2  phd=3
# Experience:   junior=0       senior=1    lead=2
# Satisfaction: low=0          medium=1    high=2
````

````python
#📌 6. Target Guided Ordinal Encoding
import pandas as pd
import numpy as np

data = {
    'Name':       ['Alice','Bob','Charlie','David','Eve',
                   'Frank','Grace','Hank','Ivy','Jack'],
    'Job_Title':  ['Cashier','Data Scientist','Analyst','Manager','Sales Exec',
                   'Analyst','Manager','Cashier','Data Scientist','Sales Exec'],
    'City':       ['Mumbai','Delhi','Mumbai','Chennai','Delhi',
                   'Kolkata','Mumbai','Delhi','Chennai','Kolkata'],
    'Purchased':  ['Yes','No','Yes','Yes','No',
                   'Yes','No','Yes','No','Yes'],
    'Salary':     [18000, 95000, 52000, 75000, 35000,
                   55000, 78000, 20000, 92000, 38000]
}
df = pd.DataFrame(data)
df
````

````python
#Step 1 — Calculate Mean Salary per Job Title
mean_salary_by_job = df.groupby("Job_Title")["Salary"].mean()
print(mean_salary_by_job)
````

````python
"""
How groupby() works:

df.groupby("Job_Title")["Salary"].mean()
```
```
groupby("Job_Title")  →  splits the DataFrame into groups:

Group 1 — Cashier:
   Alice   18000
   Hank    20000

Group 2 — Data Scientist:
   Bob     95000
   Ivy     92000

Group 3 — Analyst:
   Charlie 52000
   Frank   55000
...
```
```
["Salary"]  →  select only Salary column from each group
.mean()     →  calculate mean of Salary for each group
```

**Result:**
```
Job_Title
Analyst           53500.0
Cashier           19000.0
Data Scientist    93500.0
Manager           76500.0
Sales Exec        36500.0
"""
````

````python
#Step 2 — Create Ordinal Rank Based on Mean
# sort by mean salary and assign rank 0, 1, 2, 3, 4
mean_salary_sorted = mean_salary_by_job.sort_values()
job_title_rank = {job: rank for rank, job in enumerate(mean_salary_sorted.index)}
print(job_title_rank)

""" 
Breaking down enumerate():
mean_salary_sorted.index = ['Cashier', 'Sales Exec', 'Analyst', 'Manager', 'Data Scientist']
#                            lowest mean                                      highest mean

enumerate(mean_salary_sorted.index)
# gives: (0, 'Cashier'), (1, 'Sales Exec'), (2, 'Analyst'), (3, 'Manager'), (4, 'Data Scientist')
#         rank  job                rank  job
"""
````

````python
# same as writing:
job_title_rank = {}
for rank, job in enumerate(mean_salary_sorted.index):
    job_title_rank[job] = rank
````

````python
#Step 3 — Map the Rank Back to the Column
df["Job_Title_encoded"] = df["Job_Title"].map(job_title_rank)
df[["Job_Title", "Salary", "Job_Title_encoded"]]

""" 
How .map() works:

df["Job_Title"].map(job_title_rank)
```
Goes through every value in `Job_Title` column, looks it up in the dictionary, and replaces it with the corresponding rank:
```
Alice   →  Cashier         →  look up in dict  →  0
Bob     →  Data Scientist  →  look up in dict  →  4
Charlie →  Analyst         →  look up in dict  →  2
David   →  Manager         →  look up in dict  →  3
Eve     →  Sales Exec      →  look up in dict  →  1
...
```

**Result:**
```
     Job_Title   Salary  Job_Title_encoded
0      Cashier    18000                  0
1   Data Sci...   95000                  4
2      Analyst    52000                  2
3      Manager    75000                  3
4   Sales Exec    35000                  1
5      Analyst    55000                  2
6      Manager    78000                  3
7      Cashier    20000                  0
8   Data Sci...   92000                  4
9   Sales Exec    38000                  1
"""
````

````python
from sklearn.preprocessing import OneHotEncoder, LabelEncoder, OrdinalEncoder

df_processed = df.copy()

# 1 — Target Guided Ordinal Encoding on Job_Title (high cardinality, nominal)
mean_by_job   = df_processed.groupby("Job_Title")["Salary"].mean().sort_values()
job_rank      = {job: rank for rank, job in enumerate(mean_by_job.index)}
df_processed["Job_Title_encoded"] = df_processed["Job_Title"].map(job_rank)

# 2 — One Hot Encoding on City (low cardinality, nominal)
ohe = OneHotEncoder(sparse_output=False, drop="first")
city_encoded = ohe.fit_transform(df_processed[["City"]])
city_df = pd.DataFrame(city_encoded, columns=ohe.get_feature_names_out())

# 3 — Label Encoding on Purchased (binary)
le = LabelEncoder()
df_processed["Purchased_encoded"] = le.fit_transform(df_processed["Purchased"])

# 4 — Drop original columns, join all encoded
df_processed = df_processed.drop(columns=["Job_Title", "City", "Purchased", "Name"])
df_final = pd.concat([df_processed, city_df], axis=1)
df_final
````
