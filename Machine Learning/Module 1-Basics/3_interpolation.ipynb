```python
import pandas as pd
```

```python
# data interpo;ation 
# it is the process of estimating unknown values that fall between known values in a data set.

# Types 
# 1. Linear Interpolation: This method assumes that the change between two known values is linear and estimates the unknown value accordingly.
# 2. Polynomial Interpolation: This method fits a polynomial function to the known data points and uses it to estimate the unknown value.
#3. Cubic Spline Interpolation: This method fits a piecewise cubic function to the known data points, ensuring smoothness at the boundaries between segments.
```

```python
x = np.array([1, 2, 3, 4, 5])
y = np.array([1,3,5,7,9])

import matplotlib.pyplot as plt

plt.scatter(x, y)
plt.show()
```

```python
# linear interpolation
x_new = np.linspace(1,5,10)
y_interpolated = np.interp(x_new, x, y)
```

```python
print("New x values:", x_new)
print("Interpolated y values:", y_interpolated)
plt.scatter(x, y, label='Known Points')
plt.scatter(x_new, y_interpolated, color='red', label='Interpolated Points')
plt.legend()
plt.show()
```

Cubic Interpolation

```python
x = np.array([1, 2, 3, 4, 5])
y = np.array([1,8,27,64,125])   # y = x^3

plt.scatter(x, y)
```

```python
from scipy.interpolate import interp1d
f = interp1d(x, y, kind="cubic")

x_new = np.linspace(1,5,10)
y_interpolated = f(x_new)

print("New x values:", x_new)
print("Interpolated y values:", y_interpolated) 
```

```python
plt.scatter(x, y, label='Known Points')
plt.scatter(x_new, y_interpolated, color='red', label='Interpolated Points')
plt.legend()
```

Polynomial Interpolation 

```python
# polynomial interpolation
x = np.array([1, 2, 3, 4, 5])
y = np.array([1, 4, 9, 1, 25])
plt.scatter(x, y)
```

```python
x_new = np.linspace(1,5,10)
y_interpolated = np.polyval(np.polyfit(x, y, deg=3), x_new)
plt.scatter(x, y, label='Known Points')
plt.scatter(x_new, y_interpolated, color='red', label='Interpolated Points')
plt.legend()
```
