```python
# so how to write myltiple line : 
# method   1 
hello = open("hello.txt", "w")
hello.write("this is my first line \n")
hello.write("this is my second line \n")
hello.write("this is my third line \n")
hello.close()

# now let say you have to add one more line , 
# if you will use "w" all the previous content will get deleated >> so we will use append "a"
hello = open("hello.txt", "a")
hello.write("this is fourth line\n")
hello.close()
```

```python
# now if you want to read the file content 
# use for loop 
hello = open("hello.txt", "r")
for i in hello:
    print(i)
```

```python
# continuation of the above :
# now let say if you execute 
hello.read()
# it throw the empty string : '' >> because at this time our cursor(pointer) is at last position
```

```python
# if you want to see the position of the pointer : 
print(hello.tell() ) # cursor is at lase position 

hello.seek(0) # now cursor is at starting point i.e 0 
print(hello.tell())
```

```python
# .read() -> read the whole content of the file 

hello = open("hello.txt", "r")
hello.seek(0)
print(hello.read())
hello.close()
```

```python
# .readline() -> read the single line of the file 
hello = open("hello.txt", "r")
print(hello.readline())
print(hello.tell())
print(hello.readline())
print(hello.tell())
print(hello.readline())
print(hello.tell())


hello.close()
```

```python
# .readlines -> all the lines will become the element of the list 
hello = open("hello.txt", "r")
print(hello.readline())
hello.seek(0)
print(hello.readlines())
```
