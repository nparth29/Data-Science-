```python
# write  txt file 
with open("example.txt", "w") as file: 
    file.write(text_content)
    
# read txt file 
with open("example.txt", "r") as file:
    data = file.read()
print(data)
```

```python
# CSV -> comma seperated values files 
data = [["Name","Course","Fee"],
        ["Parth","DS","16000"],
        ["Rahul","DA","20000"]
        ]

for i in data:
    print(i)
```

```python
# now lets use the "data" to generate csv file 
import csv
with open("csvExample.csv", "w", newline='') as f:
    w = csv.writer(f)
    for i in data:
        w.writerow(i)
        
```

```python
# read the csv file 
with open("csvExample.csv", "r") as f:
    r = csv.reader(f)
    for i in r:
        print(i)
```

```python
# Write binary data to a file
with open("test.bin", "wb") as f:   # 'wb' = write binary
    f.write(b"Hello, World!")       # prefix b makes it binary data

# Read binary data from the file
with open("test.bin", "rb") as f:   # 'rb' = read binary
    data = f.read()
    print(data)                     # prints: b'Hello, World!'
    print(data.decode())            # prints: Hello, World!
```

```python
# now lets understand Buffered reading and writing 
# it allow data to read and write in chunks 
import io
with open("fileBuffer.txt", "wb") as f:
    file = io.BufferedWriter(f)
    file.write(b"Data science is a multidisciplinary field that extracts knowledge and insights from data by combining statistics, computer science, artificial intelligence, and domain expertise to solve complex problems and drive informed decisions. It involves data processing, analysis, and visualization to identify patterns, build predictive models, and uncover actionable insights from both structured and unstructured data. The field's importance is growing rapidly as the volume of data increases, leading to high demand for data scientists and their skills across various industries. ")
    file.write(b"I am studying data science. ")
    file.write(b"This is a test sample to undestand the concept of buffered reading and writing.")
    file.flush()
```
