```python
# best practics : always catch specific error
#specificity in exceptional handling 
```

```python
try:
    print(10/"hello")
except TypeError as e:
    print("String is being used in denominator : ")
    print(e)
    
    
```

```python
# Benifit of handling specific error >> you know what error is expected and if 
# other error occure which you were not able to think , you be able to address that 
```

```python
try:
    print(10/"5")
except TypeError as e:
    print("String is being used in denominator : ")
    print(e)
    print("the value should be :",10,"/",int("5"),"that is :", 10/int("5"))
    
    
```

```python
#example 2 :
try:
    print(10/0)
except ZeroDivisionError as e:
    print("Denominator must not be zero")
    print("Error: ",e)
    print("Using the default value instead of 0 that is 1 :",10/1)
```

```python
# we can also handle multiple exception :
try:
    10/0
except (ZeroDivisionError, TypeError) as e:
    print(e)
```

```python
# we can also handle multiple exception :
try:
    10/"string"
except (TypeError) as e:
    print("this is type error :",e)
except ZeroDivisionError as e:
    print("this is zero division error",e)
```

```python
# sometimes you don't know what could be the possibel exception 
try:
    f = open("exampl.txt", "r")
    f.read()
except Exception as e:  # parent class 
    print("first line -->> file not found due to :",e)
except FileNotFoundError as e:
    print("second line ---> file was not created ", e)
```

```python
# in case you know specific exceptions keep that block first
# and then block of generic exception after that 
try:
    f = open("exampl.txt", "r")
    f.read()
except FileNotFoundError as e:
    print("first line ---> file was not created ", e)
    
except Exception as e:  # parent class 
    print("second line -->> file not found due to :",e)
    
    
# so the best practice is >>>> try to write specific exceptions first and the general exception 
```

```python
# another best practics is always try to log 
import logging
logging.basicConfig(filename= "program.log", level= logging.DEBUG)

try: 
    f = open("example.txt", "r")
    f.read()
except FileNotFoundError as e:
    logging.error(f"my file was not found {e}")
```

```python
# putting exceptional handling inside function 
def test(file):
    try:
        f = open(file, "r")
        f.read()
    except FileNotFoundError as e:
        print(f"file is not found in cwd {e}")
```

```python
test("example1.txt")
```
