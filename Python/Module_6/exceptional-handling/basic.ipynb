```python
# # syntax : 
# try:
#     # suspicious code 
# except:
#     # executed only when exception occure in try block 
# except is the base class where all type of excaption is writen 
```

```python
try:
    with open("example.txt", "r") as f:
        data = f.read()
        print(data)
except Exception as e:
    print("Exception :",e)
print("this line is out of try block")
```

```python
# try except and else statement : 
# for come case you wnat to execute the line only when the code inside the try block get executed : 
#    for this purpose we use else statement 

try:
    with open("example.txt", "r") as f:
        data = f.read()
        print(data)
except Exception as e:
    print("Exception :",e)
else:
    print("this line only execute when try block get executed ")
```

```python
#example 2 
try: 
    f = open("example.txt", "w")
    f.write("this line is written inside try block ( cell 12 of notebook)")
except Exception as e:
    print("exception : ",e)
else:
    f.close() # file is closed in else block 
    print("file is written [End]")
```

```python
# try , except , else and finally 
#example 2 
try: 
    f = open("example.txt", "a")
    f.write("appended second line using 'a' mode ")
    f.write("code inside fially will always run!!")
except Exception as e:
    print("exception : ",e)
else:
    f.close() # file is closed in else block 
    print("this is else block code ")
finally: 
    print("this is finnaly block code ")
```

```python
# custom exception class : 
class validateSalary(Exception):  
    def __init__(self, msg):
        self.msg = msg
```

```python
def validate_salary(salary):
    if(salary < 0): 
        raise validateSalary("salary can not be negative ")
    elif(salary > 20000000): 
        raise validateSalary("salary is unexpected")
    else: print("salary is valid") 
```

```python
try:
    salary = int(input("Enter salary ammount :"))
    validate_salary(salary)
except validateSalary as e:  # here validateSalary is your build in exception class 
    print(e)
```
