```python
# 0. NOSET : level is not used 
# 1. DEBUG>> lowest level of logging. To give detailed info of any variable, msg while debugging 
# 2. INFO>> this is used to convey that code is working as expecting 
# example : "Data analysis is completed"
# 3. WARNING>> used to indicate that something unexpected happend or potential issue in the code 
# 4. ERROR>> Serious problem with some function 
# 5. CRITICAL >> termination of program/cide/software

# # the above level of logging is executed in the same order. 
# debug is lowest priority and ciritical is highest priority 

# while defining in logging.basicConfig, from that level the logs will be logged or in other words 
    # log priority will be start from that level which is used to initialized the logging 
    # (in our case it is defined as logging.INFO) => so debug priority is lower than INFO so it is not visible in test.log file 
```

```python
main_list = [4,[9,"orange"],"mango",78,["potato",67]]

# separate the int and str into two different list 

```

```python
#solution : 
import logging

# configure logging
logging.basicConfig(filename="list_separator.log",
                    level=logging.DEBUG,
                    format="%(asctime)s - %(levelname)s - %(message)s")

l1_int = []
l2_str = []

for i in main_list:
    logging.info(f"processing each item {i}")
    
    if type(i) == list:
        for j in i:
            logging.info(f"processing sub list {j}")
            if type(j) == int:
                l1_int.append(j)
            else :l2_str.append(j)
    elif type(i) == int:
        l1_int.append(i)
    else: l2_str.append(i)
    
print(l1_int, " ", l2_str)
logging.shutdown() # to close the log file 
```
