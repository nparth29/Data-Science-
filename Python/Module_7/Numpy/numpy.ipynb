```python
import numpy as np

# Create a 2D NumPy array
arr = np.array([[1, 2, 3], 
                [4, 5, 6]])

# Print the array
print("Array:")
print(arr)

# Key Properties
print("\nKey Properties:")
print("Dimensions (ndim):", arr.ndim)     # 2
print("Shape:", arr.shape)                # (2, 3)
print("Size:", arr.size)                  # 6
print("Data type:", arr.dtype)            # int64 or int32

# Accessing Elements
print("\nAccessing Elements:")
print("Element at [0,1]:", arr[0, 1])     # 2
print("Second row:", arr[1, :])           # [4 5 6]
print("Third column:", arr[:, 2])         # [3 6])
```

```python
# more way to convert array :
l1 = [1,2,3,4,5]
arr2 = np.asanyarray(l1)
print(arr2)
print(type(arr2))


l3 = [1,2,3,4,5]
arr3 = np.asanyarray(l3)
print(arr3)
print(type(arr3))
```

```python
"""
NOTE: asanyarray will convert the input to an array but if something is already in anrray it will not be able to convert but pass ndarray subclass through
 therefore if something is already in array or part of array it will simpally allow to pass through it 
"""     
np.asanyarray(arr3)
```

```python
""" shallow copy """

l = [1,2,3,4,5]
arr = np.array(l)
print(f"arr: {arr}")

brr = arr # brr and arr will point to the same memory location, any changes done to 'arr' or 'brr'  will also reflect on other.
print("brr:", brr)

arr[0] = 100

print(f" arr: {arr},     brr: {brr}")
```

```python
"""deep copy"""
crr = arr.copy()
print(f"crr: {crr}")

arr[1] = 200
crr[2] = 300
print(f"arr: {arr}")
print(f"brr: {brr}")
print(f"crr:{crr}")
```

```python
"""Creating array from np.fromfunction()"""
arr = np.fromfunction(lambda i,j: i==j, (3,3))
print(f"arr: {arr}")
print(arr.ndim)
print(arr.size)
print(arr.shape)
```

```python
# Function that just returns row + col
def func(i, j):
    return i + j

arr = np.fromfunction(func, (3, 3), dtype=int)
print(f"arr: {arr}")
```

```python
import numpy as np

# np.fromstring() converts a string into a 1D NumPy array.
# Arguments:
#  - The first argument is the input string.
#  - sep specifies how values in the string are separated (delimiter).

d = np.fromstring("22 23 24", sep=" ")

# Print the array
print(d)
```

```python
import numpy as np

# Define a string
string = "Ajay, Vijay, Sanjay"

# Use Python's split() method to break the string into parts
# split(",") means "split wherever a comma occurs"
parts = string.split(",")

# Convert the resulting Python list into a NumPy array
crr = np.array(parts)

# Print the NumPy array
print(crr)
```
