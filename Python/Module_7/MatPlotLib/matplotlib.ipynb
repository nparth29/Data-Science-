### Plotting Multiple Lines with Labels

To plot multiple lines, simply call `plt.plot()` for each line you want to add to the same figure. Use the `label` parameter for each line, and then call `plt.legend()` to display these labels.

```python
x1_line = [1, 2, 3, 4, 5]
y1_line = [5, 2, 7, 8, 1]

x2_line = [1, 2, 3, 4, 5]
y2_line = [1, 8, 3, 2, 9]

plt.plot(x1_line, y1_line, color = 'red', marker = "o", linestyle = "--", label = "Stock A Price")
plt.plot(x2_line, y2_line, color = 'blue', marker = "x", linestyle = "-", label = "Stock B Price")

plt.title("Multiple Line Chart with Labels")
plt.xlabel("Days")
plt.ylabel("Stock Price")
plt.grid(True)
plt.legend() # This displays the labels
plt.show()
```

```python
x = np.linspace(0,10,100) # will generate a NumPy array containing 100 numbers, starting from 0 and ending at 10, with each number being equidistant from its neighbors.
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
```

```python
plt.figure() # to initialize a page to make a plot
plt.subplot(1,2,1)
plt.plot(x, np.sin(x))

plt.subplot(1,2,2)
plt.plot(x, np.cos(x))

plt.show()

```

plt.figure(): This line creates a new figure (think of it as a blank canvas or a page) to draw your plots on. It's a good practice to start with this when you want more control over your plots, especially if you're creating multiple plots.

plt.subplot(1, 2, 1): This is the key part for creating multiple plots within the same figure. subplot() divides the figure into a grid, and then you select which part of the grid to plot on.

The first number (1) means you want 1 row of plots.
The second number (2) means you want 2 columns of plots.
The third number (1) means you are selecting the first plot position in this 1x2 grid.
plt.plot(x, np.sin(x)): This line then plots the sine wave (np.sin(x)) against the x values in the first subplot position you just selected.

plt.subplot(1, 2, 2): Similar to the first subplot call, this again defines a 1x2 grid, but this time you are selecting the second plot position.

plt.plot(x, np.cos(x)): This line plots the cosine wave (np.cos(x)) against the x values in the second subplot position.

plt.show(): Finally, this displays the entire figure with both subplots. If you didn't include this, the plots might not render or might show some metadata along with the plot.

```python
plt.figure() # to initialize a page to make a plot
plt.subplot(2,1,1)
plt.plot(x, np.sin(x))

plt.subplot(2,1,2)
plt.plot(x, np.cos(x))

plt.show()

```

```python
import matplotlib.pyplot as plt

# Data for the pie chart
# These are the numerical values that determine the size of each slice.
sizes = [15, 30, 45, 10]

# Labels for each slice of the pie chart.
# The order of labels corresponds to the order of sizes.
labels = ['Apples', 'Bananas', 'Cherries', 'Dates']

# Colors for each slice. Optional, if not provided, matplotlib chooses defaults.
colors = ['red', 'yellow', 'lightcoral', 'lightskyblue']

# Optional: How to display the percentage value on each slice.
# '%1.1f%%' means format as a float with 1 decimal place, followed by a percent sign.
# For example, 15.0% or 30.5%
autopct = '%1.1f%%'

# Optional: Specifies the starting angle of the first wedge. Default is 0.
# Here, we start the first slice (Apples) at the 90-degree position (top).
startangle = 90

# Create the pie chart
# 'explode' is an optional argument to make a slice 'pop out' from the pie.
# Here, we define a tuple for 'explode' where 0 means no explosion, and a value like 0.1 means explode by 10% of the radius.
explode = (0, 0.1, 0, 0) # Only "explode" the 2nd slice (Bananas)

plt.figure(figsize=(8, 8))
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct=autopct, shadow=True, startangle=startangle)

plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Fruit Distribution Example')
plt.show()
```

Let's break down the components of the `plt.pie()` function and the keywords used:

*   **`sizes`**: This is a list or array of numerical values representing the **wedge sizes**. Each value corresponds to a slice of the pie.

*   **`labels`**: This is a list of strings that provide **text labels** for each wedge. These labels are displayed next to their corresponding slices.

*   **`colors`**: (Optional) A list of color names or hex codes to assign specific colors to each wedge. If not provided, Matplotlib will cycle through its default color palette.

*   **`autopct`**: This is a **format string** or a function used to display the percentage value within each slice.
    *   `'%1.1f%%'` is a common format string that means:
        *   `%`: Marks the start of a format specifier.
        *   `1.1f`: Formats a floating-point number with a minimum width of 1 and **1 digit after the decimal point**.
        *   `%%`: Displays a literal percent sign.

*   **`startangle`**: This parameter specifies the **angle at which the first wedge starts**, measured counterclockwise from the x-axis. For example, `startangle=90` makes the first slice begin at the top of the circle.

*   **`explode`**: (Optional) A tuple or array that controls how much each wedge is "exploded" or separated from the center of the pie. It contains a fraction of the radius for each wedge. A value of `0` means the wedge is attached to the center, while a value like `0.1` means it's offset by 10% of the radius.

*   **`shadow`**: (Optional) If set to `True`, it adds a shadow effect beneath the pie chart, giving it a 3D appearance.

*   **`plt.axis('equal')`**: This ensures that the pie chart is drawn as a perfect circle rather than an ellipse, which can happen if the aspect ratio of the plot is not equal.

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for the stackplot
# x-axis values (e.g., time points or categories)
x = np.arange(1, 10, 1)

# y-axis values for different series/components
# Each row represents a series, and each column corresponds to an x-value.
y1 = np.array([1, 2, 3, 4, 3, 2, 1, 2, 3])
y2 = np.array([2, 3, 4, 5, 4, 3, 2, 3, 4])
y3 = np.array([3, 4, 5, 6, 5, 4, 3, 4, 5])

# You can also generate random data for demonstration
# x = np.arange(1, 11)
# y1 = np.random.randint(1, 10, size=10)
# y2 = np.random.randint(1, 10, size=10)
# y3 = np.random.randint(1, 10, size=10)

# Stackplot function
# It takes x and multiple y arrays as input.
# The 'labels' argument is used for the legend, and 'colors' sets the color for each stack.
plt.stackplot(x, y1, y2, y3, labels=['Series 1', 'Series 2', 'Series 3'], colors=['skyblue', 'lightcoral', 'lightgreen'])

# Adding plot enhancements
plt.xlabel('X-axis (Time/Category)')
plt.ylabel('Y-axis (Value)')
plt.title('Example of a Stackplot (Area Plot)')
plt.legend(loc='upper left') # Displays the labels for each series
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```
