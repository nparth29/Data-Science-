```python

series_one = pd.Series(list(df1['age'][2:5]), index=['a', 'b', 'c'])
series_two = pd.Series(list(df1['age'][5:8]), index = ['row5', 'row6', 'row7'])
print(series_one)
print("\n")
print(series_two)

# concatenate two series: >> series with different indexes and dataferame can not be concatenated directly using '+'
# if you keep the indexes same then it will work
series_three = pd.Series(list(df1['age'][2:5]), index=['row5', 'row6', 'row7'])
print("\n")
print(series_three)
print("\n")
print(series_two + series_three)  # it will add values of same index and for different index it will give NaN


# DOUBT: Why df1['age'][2:5] and not df1[2:5]['age'] ?
# SOL: Both work. df1['age'][2:5] means — get age column first, then slice rows 2,3,4 from it.
#      df1[2:5]['age'] means — slice rows first, then get age column. Same result.

# DOUBT: Why no loc here ?
# SOL: loc is for precise row+column selection together. Here we just need a quick slice of one column.
#      loc equivalent would be >> df1.loc[2:4, 'age'] — same result but more verbose.

# DOUBT: Why list() is used ?
# SOL: df1['age'][2:5] returns a Series (with index 2,3,4).
#      list() converts it to a plain Python list [26.0, 35.0, 35.0]
#      so that when we create a new Series with custom index ['a','b','c'], there is no index conflict.

# DOUBT: Why df1['age'][2:5] is considered chained indexing and is it bad ?
# SOL: Yes, [][] is called chained indexing. It can cause SettingWithCopyWarning in some cases.
#      For reading data it is fine, but for modifying data always use loc >> df1.loc[2:4, 'age'] = value
```

```python
#drop rows and columns
# by default if you drop anything pandas assume you are dropping rows axis=0

import pandas as pd
df2 = pd.read_csv("titanic.csv")
print(df2.head())
print("\n")

# lets say we have to drop 'who' column 
# df2.drop("who") => gives you error because by default it will try to drop row with index 'who' which is not present
# to drop 'who' column we have to do axix = 1 
df2.drop("who" , axis = 1 , inplace = True)  # inplace = True means changes will be made in original dataframe
# so => axis = 0 >> rows
#    => axis = 1 >> columns
# inplace = True >> changes will be made in original dataframe
print(df2.head())
```

```python
# to drop a particular row
df2.drop(0 , axis = 0 , inplace = True)  # dropping first row

```

```python
# how to see null values in dataframe
df4 = pd.read_csv("titanic.csv")
print(df4.isnull().sum()) # gives count of null values in each column
# so either we can drop those null values or we can fill them with some value ( imputation )

# 1. Dropping null values
print(f"before dropping null values: shape : {df4.shape}")
df4.dropna(inplace = True)  # drops all rows which have at least one null value
print(df4.isnull().sum())  # now it will show 0 null values in
print(f"after dropping null values: shape : {df4.shape}")
```

```python
# if we put axis = 1 then it will drop all columns which have at least one null value
df5 = pd.read_csv("titanic.csv")
print(f"before dropping null values: shape : {df5.shape}")
df5.dropna(axis = 1 , inplace = True)  # drops all columns which have at least one null value
print(f"after dropping null values: shape : {df5.shape}")

```

```python
#imputation of missing values >> replacing mussing values with some value
#numberical columns >> we can replace missing values with mean , median
# before that we have to remove outliers then only mean and median will be useful


#categorical columns >> we can replace missing values with mode (most frequent value) mode
df7 = pd.read_csv("titanic.csv")
print(df7.isnull().sum())
```

```python
# replacing missing value with some conastant value")
#df7['embarked'].fillna('S' , inplace = True)  # replacing missing values in 'embarked' column with 'S'
df7.fillna(0, inplace=True)
print(df7.isnull().sum())
```

```python
df8 = pd.read_csv("Sample_null_value.csv")
print(df8)

# df8['Age'].fillna(df8['Age'].mean() , inplace = True) # this is the older verson
df8.fillna({
    'Age': df8['Age'].mean(),             # replace missing Age with mean
    'Salary': df8['Salary'].median(),     # replace missing Salary with median
    'Department': df8['Department'].mode()[0]  # replace missing Department with mode (most frequent)
}, inplace=True)

print(df8)
```

```python
#ffill concept : it says to fill the missing value with previous value in that column
df9 = pd.read_csv("Sample_null_value.csv")
print(df9)
print("\n\nAfter applying ffill method\n")
df9.fillna(method = 'ffill' , inplace = True)
print(df9)

# NOTE: If the first value(s) in a column are NaN, ffill will not fill them since there is no previous value to carry forward.
```

```python
# bfill concept : it says to fill the missing value with next value in that column
df10 = pd.read_csv("Sample_null_value.csv") 
print("\n\nBefore applying bfill method\n")
print(df10)
df10.fillna(method = 'bfill' , inplace = True)
print("\n\nAfter applying bfill method\n")
print(df10)
#NOTE: If the last value(s) in a column are NaN, bfill will not fill them since there is no next value to carry backward.
```

```python
# Que 1: what is the average fare paid by the people who survived ? titanic.csv
df11 = pd.read_csv("titanic.csv")
print(df11.head(3))
print("\n")

print(df11[df11['survived'] == 1].head(3))

avg_fare = df11[df11['survived'] == 1]['fare'].mean()
print(f"The average fare paid by the people who survived is: {avg_fare}")
```

```python
#  we will be use groupby function

print(df11.groupby('survived').mean(numeric_only=True))
# working of groupby function:
# first it will divide the dataframe into two parts where survived = 0 and survived = 1
# then it will calculate mean of each column for both parts and give the result
```

```python
#lets use aggrigate function with groupby
import pandas as pd
import numpy as np
df12 = pd.read_csv("titanic.csv")
print(df12.groupby('survived').agg({
    'age': ['mean', 'median', 'min', 'max'],
    'fare': ['mean', 'median', 'min', 'max'],
    'pclass': 'count'
}))

# for std : np.std will be used
# print(df12.groupby('survived').agg({
#     'age': np.std,
#     'fare': np.std
# }))
```

```python
# how many passengers survived / not survived for each sex and pclass 
df13 = pd.read_csv("titanic.csv") 
print(df13.groupby(['sex', 'pclass', 'survived']).size())

# how groupby works here:
# first it will divide the dataframe into multiple parts based on unique combinations of 'sex' , 'pclass' and 'survived'
# then it will count number of rows in each part and give the result   
```

```python
# convert the above groupby result into a dataframe
df_grouped = df13.groupby(['sex', 'pclass', 'survived']).size().reset_index(name='count')
print(df_grouped)
```

```python
# we can also use to_frame 
df_grouped = df13.groupby(['sex', 'pclass', 'survived']).size().to_frame(name='count').reset_index()
print(df_grouped)
```
