```python
import pandas as pd
df1 = pd.read_csv("titanic.csv")
print(df1.head())
```

```python
# accessing rows and columns of dataframe >> implicit/integer/internal indexing  explecit or named indexing
```

```python
#Implesit indexing
#iloc[row_indexer , column_indexer]
df1.iloc[0:5 , 0:3]  # first 5 rows and first 3 columns
```

```python
# Explicit indexing
#loc[row_indexer , column_indexer]
print(df1.loc[0:4 , ["age" , "fare"]])  # first 5 rows and age , fare columns
print("\n \n \n")
print(list(df1.loc[0:4 , ["age" , "fare"]] .itertuples()))
```

```python

series_one = pd.Series(list(df1['age'][2:5]), index=['a', 'b', 'c'])
series_two = pd.Series(list(df1['age'][5:8]), index = ['row5', 'row6', 'row7'])
print(series_one)
print("\n")
print(series_two)

# concatenate two series: >> series with different indexes and dataferame can not be concatenated directly using '+'
# if you keep the indexes same then it will work
series_three = pd.Series(list(df1['age'][2:5]), index=['row5', 'row6', 'row7'])
print("\n")
print(series_three)
print("\n")
print(series_two + series_three)  # it will add values of same index and for different index it will give NaN
```

```python
#drop rows and columns
# by default if you drop anything pandas assume you are dropping rows axis=0

import pandas as pd
df2 = pd.read_csv("titanic.csv")
print(df2.head())
print("\n")

# lets say we have to drop 'who' column 
# df2.drop("who") => gives you error because by default it will try to drop row with index 'who' which is not present
# to drop 'who' column we have to do axix = 1 
df2.drop("who" , axis = 1 , inplace = True)  # inplace = True means changes will be made in original dataframe
# so => axis = 0 >> rows
#    => axis = 1 >> columns
# inplace = True >> changes will be made in original dataframe
print(df2.head())
```

```python
# to drop a particular row
df2.drop(0 , axis = 0 , inplace = True)  # dropping first row

```

```python
# how to set a new index 
df3 = pd.read_csv("titanic.csv")
print(df3.head(3))
print("\n")
df3.set_index("age" , inplace = True)
print(df3.head(3))

''' 
      survived  pclass     sex  sibsp  parch     fare embarked  class    who  \
age                                                                            
22.0         0       3    male      1      0   7.2500        S  Third    man   
38.0         1     1    female      1      0  71.2833        C  First       woman


Now > row1 is known by index 22.0 and row2 is known by index 38.0
'''
# lets access row with index 38.0
print(df3.loc[38.0])
```

```python
# continue of block 6 : 
# now lets reset the index to default integer index
df3.reset_index(inplace = True)
print(df3.head(3))
```
