rolling function 

```python
df4 = pd.DataFrame({
    'a': [1,2,3,4,5,6,7,8,9]
})
df4
```

```python
df4.rolling(window=2).mean()  # here window=3 means it will take 2 values at a time and calculate mean
```

```python
df4.rolling(window=3).sum() # here window=3 means it will take 3 values at a time and calculate sum
# first 2 values will be NaN because there are not enough values to fill the window of 3 
```

```python
df4.rolling(window=3).sum().dropna() # to drop NaN values
```

```python
#typecasting 
df4['a_float'] = df4['a'].astype(float)
df4 
```

```python
import numpy as np
df_random = pd.DataFrame({
    'a': [1,2,3,4,5,6,7,8,9],
    'b': [4,3,7,5,2,8,1,9,6]
})
df_random
```

```python
df_random['group'] = np.random.choice(['a','b'], size=9)
df_random
```

```python

df_random.groupby('group')['a'].sum() 
''' 
✅ 1) groupby('group')

Splits the DataFrame into separate mini-tables based on the values in the "group" column.

If "group" contains 'a' and 'b', you now have:

one group containing all rows where group == 'a'

one group containing all rows where group == 'b'

✅ 2) ['a']

From each group, select only column 'a'.
Column 'b' is ignored completely.

✅ 3) .sum()

For each group, add all the values in column 'a'.
'''
```

```python
#let say we have to do rolling sum on groupby
df_random.groupby('group')['a'].rolling(window=2).sum()
    
```

```python
# cumulative sum 
df_random.a.cumsum()
```
