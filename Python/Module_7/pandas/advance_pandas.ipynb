```python
import pandas as pd

# Load Titanic dataset
df_titanic = pd.read_csv("titanic.csv")

# 1. Convert 'pclass' to categorical
print("Categorical representation of 'pclass':")
print(pd.Categorical(df_titanic["pclass"]))
print("\n")

# 2. Find unique values in 'sex' column
print("Unique values in 'sex' column:")
print(df_titanic["sex"].unique())
print("\n")

# 3. Count frequency of each class in 'pclass'
print("Count of each category in 'pclass':")
print(df_titanic["pclass"].value_counts())
print("\n")

# 4. Percentage distribution of each class
print("Percentage of each class (normalized):")
print(df_titanic["pclass"].value_counts(normalize=True) * 100)  # multiply by 100 to see %
```

```python
# lets solve some questions 
# Q1. How many persone of age less than 18 are there in the dataset ?
import pandas as pd
df = pd.read_csv("titanic.csv")
print(df[df["age"] < 18])
print("\n count of persons with age less than 18 : ", len(df[df["age"] < 18]))
#or you can use .shape 
print("\n count of persons with age less than 18 : ", df[df["age"] < 18].shape[0])
```

```python
#Que 2. Name of the passenger who is less than 10years old 
import pandas as pd
df = pd.read_csv("titanic.csv")
print(df[df["age"] < 10].Name)
#or we can use dot notation
print(df[df.age < 10].name)
```

```python
# Que 3. How many peopel paid less thand average fair 
print("\n average fair is : ", df["fare"].mean())
print(df[df["fare"] < df["fare"].mean()])
print("\n count of person who paid less than average fair: ", len(df[df["fare"] < df["fare"].mean()]))
```

```python
# Que4. How many pessengers are male
print(df[df["sex"] == "male"])
len(df[df["sex"] == "male"])
```

```python
print(df[df["pclass"] == 1])
print(len(df[df["pclass"] == 1]))
```

```python
# how many female paid more thatn average fare
avg = df["age"].mean()
print((df["sex"] ==  "female") & (df["age"] > avg))
len((df["sex"] ==  "female") & (df["age"] > avg))
```

```python
#calculate mean using numpy 
import numpy as np
mean_age = np.mean(df["age"])
print("mean age using numpy is : ", mean_age)
```

```python
# people who paid max fair 
max_fare = df["fare"].max()
print("max fare is : ", max_fare)
print(df[df["fare"] == max_fare])
```
