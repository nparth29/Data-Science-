```python
# pd.series >> 
import pandas as pd

s = pd.Series([10, 20, 30, 40])
print(s)
print(type(s))

print(s[0])  # Accessing first element
print(s[1])
s[3] = 100

print(f"modified series : \n{s}")


# we can provide index also

t = pd.Series([10, 20, 30, 40], index=['a', 'b', 'c', 'd'])
print(t)

# to see all the index : 
print(t.index)

# reset the index 
t = t.reset_index(drop=True)
print(f"reset index : \n{t}")

```

```python
# adding a new column in existing series
series_data = pd.Series([100, 200, 300, 400], index=['a', 'b', 'c', 'd'])
df = pd.DataFrame(series_data, columns=['Values'])
print(df.head())

# adding new column : 
df["new column"] = [1, 2, 3, 4 ]

print(f"\n {df.head()}")


# access a particular column : 
print(f"\n {df["new column"]}")

# lets reset the index again without drop = True >> it will add the old index as a new column
df = df.reset_index()
print(f"\n {df}")
```

```python
import requests

url = "https://github.com/datablist/sample-csv-files/raw/main/files/people/people-100.xlsx"
response = requests.get(url)

# Save the file locally in your repo
with open("people-100.xlsx", "wb") as f:
    f.write(response.content)

print("✅ Excel file downloaded and saved as people-100.xlsx")
```
