```python
import pandas as pd

data = {
    "Name": [f"Person_{i}" for i in range(1, 11)],
    "Sex": ["Male", "Female"] * 5,
    "Age": list(range(20, 30))
}

df1 = pd.DataFrame(data)
print(df1.head())
#--------------------------------
import pandas as pd

data = {
    "Name": [f"Person_{i}" for i in range(11, 21)],
    "Sex": ["Male", "Female"] * 5,
    "Age": list(range(20, 30))
}

df2 = pd.DataFrame(data)
print(df2.head())
#--------------------------------
# axis=0 -> row wise concatenation
# axis=1 -> column wise concatenation
#--------------------------------

```

```python
df3 = pd.concat([df1, df2], ignore_index=True) # ignore_index to reset index -> otherwise original index will be retained 0-9,0-9
print(df3)
```

```python
df4 = pd.concat([df1, df2], axis=1)
print("Column wise concatenation: \n")
print(df4)
```

```python
# if the index are not same
# df1 with index 0 to 4
df1 = pd.DataFrame({
    "A": [1, 2, 3, 4, 5]
})

# df2 with index 10 to 14
df2 = pd.DataFrame({
    "A": [100, 200, 300, 400, 500]
}, index=[10, 11, 12, 13, 14])

print("DF1:")
print(df1)
print("\nDF2:")
print(df2)

# Horizontal concat (axis=1)
df_h = pd.concat([df1, df2], axis=1)

print("\nHorizontal Concat (axis=1):")
print(df_h)

# now lets reset index 
df_pos = pd.concat([df1.reset_index(drop=True), df2.reset_index(drop=True)], axis=1)
print(df_pos)
```

Merging Datasets 


```python
# we are using vein diagram to understand merging
# let say we have two datasets : A and B
# 1.    merging based on common columns -> inner merge (intersection of A and B)
#2.     merging based on left dataset(A) -> left merge (all of A and common of B)
# 3.    merging based on right dataset(B) -> right merge (all of B and common of A)
# 4.    merging based on all values from both datasets -> outer merge (union of A and B)
# 5.    cross merge -> cartesian product of A and B (every combination of A and B)
```

```python
df_merge1 = pd.DataFrame({
    "key1": [1,2,4,5,6],
    "key2": [4,5,6,7,8],
    "key3": [3,4,5,6,6]
})
print("DF1:")
print(df_merge1)

df_merge2 = pd.DataFrame({
    "key1": [1,2,45,6,67],
    "key4": [56,5,6,7,8],
    "key5": [3,56,5,6,6]
})
print("\nDF2:")
print(df_merge2)
```

```python
#inner merge
df_inner = pd.merge(df_merge1,df_merge2, how = "inner")
print("\nInner Merge:")
print(df_inner)
# how it wroks : 
""" 
key1 from df1 : 1,2,4,5,6
key1 from df2 : 1,2,45,6,67
common values are 1,2,6 -> these rows will be in final dataframe
So final dataframe will have rows with key1 as 1,2,6 and corresponding values from both dataframes.
for df_merge1:  1 <-> (4,3), 2 <-> (5,4), 6 <-> (8,6)
for df_merge2:  1 <-> (56,3), 2 <-> (5,56), 6 <-> (7,6)
Final dataframe:
   key1  key2  key3  key4  key5
0     1     4     3    56     3
1     2     5     4     5    56
2     6     8     6     7     6

"""
```

```python
# left merge
df_left = pd.merge(df_merge1,df_merge2, how = "left")
print("\nLeft Merge:")
print(df_left)
# how it works :
"""
key1 from df1 : 1,2,4,5,6
key1 from df2 : 1,2,45,6,67
all values from df1 will be in final dataframe : key1 as 1,2,4,5,6 & key2 as 4,5,6,7,8 & key3 as 3,4,5,6,6
for df2 : key1 column of df2 is matched with df1's key1 column and we have common values -> 1,2,6 
          so in final dataset the corresponding values from df2 will be added for these common key1 values 
            for other key1 values(4,5) there is no match in df2 so NaN will be added for those rows.
            1 <-> (56,3), 2 <-> (5,56), 6 <-> (7,6) and rest are NaN

Final dataframe:
   key1  key2  key3  key4  key5
0     1     4     3  56.0   3.0
1     2     5     4   5.0  56.0
2     4     6     5   NaN   NaN
3     5     7     6   NaN   NaN
4     6     8     6   7.0   6.0
"""
```

```python
# right merge
df_right = pd.merge(df_merge1,df_merge2, how = "right")
print("\nRight Merge:")
print(df_right)
# how it works :
"""
key1 from df1 : 1,2,4,5,6
key1 from df2 : 1,2,45,6,67
all values from df2 will be in final dataframe : key1 as 1,2,45,6,67 & key4 as 56,5,6,7,8 & key5 as 3,56,5,6,6
for df1 : key1 column of df1 is matched with df2's key1 column and we have common values -> 1,2,6 
          so in final dataset the corresponding values from df1 will be added for these common key1 values 
            for other key1 values(45,67) there is no match in df1 so NaN will be added for those rows.
            1 <-> (4,3), 2 <-> (5,4), 6 <-> (8,6) and rest are NaN
Final dataframe:
   key1  key2  key3  key4  key5
0     1   4.0   3.0    56     3
1     2   5.0   4.0     5    56
2    45   NaN   NaN     6     5
3     6   8.0   6.0     7     6
4    67   NaN   NaN     8     6
"""
```

```python
# outer merge
df_outer = pd.merge(df_merge1,df_merge2, how = "outer") # OUTER MEANS : all the common element of the common columns from both dataframes
print("\nOuter Merge:")
print(df_outer)
# how it works :
"""
key1 from df1 : 1,2,4,5,6
key1 from df2 : 1,2,45,6,67
common values are 1,2,6 -> these rows will have corresponding values from both dataframes
non common values are 4,5 from df1 and 45,67 from df2
for 4 and 5 of df1 -> 4 <->(6,5) and 5 <->(7,6) will have NaN for df2 columns 
for 45 and 67 of df2 -> 45 <->(6,5) and 67 <->(8,6) will have NaN for df1 columns
Final dataframe:
    key1  key2  key3  key4  key5
0     1   4.0   3.0  56.0   3.0
1     2   5.0   4.0   5.0  56.0
2     4   6.0   5.0   NaN   NaN
3     5   7.0   6.0   NaN   NaN
4     6   8.0   6.0   7.0   6.0
5    45   NaN   NaN   6.0   5.0
6    67   NaN   NaN   8.0   6.0
"""
```

```python
# cross merge
df_cross = pd.merge(df_merge1,df_merge2, how = "cross")
print("\nCross Merge:")
print(df_cross)
# how it works :
"""
Every row of df1 is combined with every row of df2
So if df1 has m rows and df2 has n rows, the resulting dataframe will have m*n rows.
If df1 has 5 rows and df2 has 5 rows, the resulting dataframe will have 5*5 = 25 rows.
"""
```

```python
# let say we have to use a specific column for merging
df_specific = pd.merge(df_merge1, df_merge2, how = 'left', left_on = 'key2', right_on = 'key4')
print("\nSpecific Column Merge (left_on='key2', right_on='key4'):")
print(df_specific)
""" 
since the merge is happend on differnt column -> so key1 from df1 represented as df1_x to specify it is from df1 and same goes for df2 
"""
```

```python
#Join > works based on index while merging works based on column values
```

```python
df_join1 = pd.DataFrame({
    "key1": [1,2,4,5,6],
    "key2": [4,5,6,7,8],
    "key3": [3,4,5,6,6]
})
# lets change the index values of df_join1
df_join1.index = ["a","b","c","d","e"]
print("DF1:")
print(df_join1)

df_join2 = pd.DataFrame({
    "key4": [1,2,45,6,67],
    "key5": [56,5,6,7,8],
    "key6": [3,56,5,6,6]
})
df_join2.index = ["a","b","f","g","h"]
print("\nDF2:")
print(df_join2)
```

```python
print(df_join1.join(df_join2, how='inner'))
print(f"\n {df_join2.join(df_join1, how='inner')}")
```

```python
print(df_join1.join(df_join2, how='left'))
print(f"\n {df_join2.join(df_join1, how='left')}")
```

```python
print(f"\n {df_join2.join(df_join1, how='right')}")
print(f"\n {df_join1.join(df_join2, how='right')}")
```
