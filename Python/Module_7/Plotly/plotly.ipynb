```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
# plotly used two things
#graph_object
#express
import plotly.graph_objects as go
import plotly.express as px
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x =[1,2,3,4,5], y = [1,7,8,2,6], mode = "markers"))
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x =[1,2,3,4,5], y = [1,7,8,2,6], mode = "lines"))
```

```python
tips = sns.load_dataset("tips")
tips
```

```python
# lets create a scatter plot useing total_bills and tip
fig = go.Figure()
fig.add_trace(go.Scatter(x = tips.total_bill, y = tips.tip, mode = "markers"))
```

```python
fig = go.Figure()
fig.add_trace(go.Bar(x = [1,2,3,4,5], y = [3,8,5,2,6]))
```

```python
# lets create a histogram
fig = go.Figure()
fig.add_trace(go.Histogram(x = tips.total_bill))
```

```python
# in scatterplot code lets change the marksr size wrt to "size" column of tips (no. of customer visit)
fig = go.Figure()
fig.add_trace(go.Scatter(x = tips.total_bill, y = tips.tip, mode = "markers", marker_size = 12*tips["size"]))
#
```

```python
from numpy._core.fromnumeric import size
#3d plot
fig = go.Figure()
fig.add_trace(go.Scatter3d(x = tips.total_bill, y = tips.tip, z = tips["size"], mode = "markers", marker_size = 10*tips["size"]))
```

```python
# lets create 3d line chart
fig = go.Figure()
fig.add_trace(go.Scatter3d(x = [1,2,3,4,5], y = [3,8,5,2,6], z = [1,2,7,3,5]))
```

Express

```python
#using plotly express as px
import plotly.express as px
x = [1,2,3,4,5]
y = [3,8,5,2,6]
fig = px.scatter(x,y)
fig.show()
```

```python
fig = px.line(x,y)
fig.show()
```

```python
#bar
x = [1,2,3,4,5]
y = [3,8,5,2,6]
fig = px.bar(x,y)
fig.show()
```

```python
# histogram
import seaborn as sns
tips = sns.load_dataset("tips")
fig = px.histogram(tips, x = "total_bill")
fig.show()
```

```python
# box plot
fig = px.box(tips, x = "day", y = "total_bill")
fig.show()
#
```

```python
data = [[1,2,3], [4,5,6],[7,8,9]]
fig = px.imshow(data)
fig.show()
#
```

```python
# bubble chart
fig = px.scatter(tips, x = "total_bill", y = "tip", size = "size")
fig.show()
```

```python
# 3d chart
fig = px.scatter_3d(tips, x = "total_bill", y = "tip", z = "size")
fig.show()
```

```python
# contour plot
x = [1,2,3,4,3,23,80]
y = [5,8,12,3,12,35,60]
px.density_contour(x,y)
```
