```python
# numerical data plotting 
# relplot has options of plotting both categorical and numerical data 
# relplot >> how is the relationship between two variables  x
```

```python
# lets see how tip is varying with total_bill
# by default relplot uses scatter plot
sns.relplot(x='total_bill', y='tip', data=df) # x and y are numerical columns, data is dataframe
plt.show()
```

```python
# let say we have to differnetiate smoking and non smoking customers
sns.relplot(x = "total_bill", y = "tip", hue = "smoker", data = df)  #hue is used to differentiate categorical data
plt.show()
# insight: for same total bill smoking customers give more tip than non smoking customers
```

```python
#hue give the different colors to the markers
# lets use style parameter to change the marker style
sns.relplot(x = "total_bill", y = "tip", style = "size", data = df)
```

```python
# hue and style can be used together
sns.relplot(x = "total_bill", y = "tip", hue = "smoker", style = "size", data = df)
plt.show()
```

```python
# relplot can also plot line plots
# line plot is used to see the trend of data over a period of time
# a time column/sequences is generally involved in line plots

time = np.arange(500) # creates an array of numbers from 0 to 499
value = np.random.randn(500).cumsum() # creates an array of 500 random numbers and takes cumulative sum

# creating a dataframe
data = pd.DataFrame({'time': time, 'value': value}) # { column_name: data_for_that_column }
#-----------------------------------------------------------------------
sns.relplot(x = "time", y = "value", kind = "line", data = data, sort = True)
plt.show()
#-----------------------------------------------------------------------
```

```python
# lets see line plot with seaborn's inbuilt dataset
df = sns.load_dataset('fmri')
df.head()
```

```python
sns.relplot(x = "timepoint", y = "signal", kind = "line", data = df, hue = "event")
plt.show()

# in graph you can see the shaded region around the line
# it represents the confidence interval
# by default seaborn shows 95% confidence interval
sns.relplot(x = "timepoint", y = "signal", kind = "line", data = df, hue = "event", ci = 'sd')
```

```python
# lets use style parameter to differentiate between different regions
sns.relplot(x = "timepoint", y = "signal", kind = "line", data = df, hue = "event", style = "region")
plt.show()
```

```python
# dashes = false will remove the dashes from the line
sns.relplot(x = "timepoint", y ="signal", kind = "line", data = df, hue = "event", style = "region", dashes = False)
plt.show()
```

```python
# put markers = True to show markers on the line
sns.relplot(x = "timepoint", y = "signal", kind = "line", data = df, hue = "event", style = "region", dashes = False, markers = True  )
plt.show()
```

```python
# relationship between day and total_bill
sns.catplot(x = "day", y = "total_bill", data = tips) # catplot is used for categorical data
plt.show()

# “Saturday and Sunday have more bill observations and higher bill values, suggesting higher restaurant activity on weekends.”
# and total_bill is also high on these days
```

```python
# in above graph we can see datapoints are overlapping
# to avoid overlapping we can use swarm plot
sns.catplot(x = "day", y = "total_bill", data = tips, kind = "swarm")
plt.show()
```

```python
# box plot to see the outliers in total_bill for each day
sns.catplot(x = "day", y = "total_bill", data = tips, kind = "box")
plt.show()
# in the graph : 
# the lower horizontal line(out of the box) represent the minimum value(lower fence)
# the upper horizontal line(out of the box) represent the maximum value(upper fence)
# the line inside the box represent the median value (Q2) -> Q represents quartiles
# Q1 and Q3 are represented by the lower and upper edges of the box respectively
# the bubbles represent the outliers 
# Q1: 25th percentile, Q2: 50th percentile(median), Q3: 75th percentile
```

```python
# lets use violin plot
sns.catplot(x = "day", y = "total_bill", data = tips, kind = "violin", color = "red")
plt.show()
```

```python
# the one more plot is : boxen plot 
sns.catplot(x = "day", y = "total_bill", data = tips, kind = "boxen")
```

```python
import seaborn as sns
tips = sns.load_dataset("tips")
# bar plot has nothing to do with count , if the bar is higher it not mean, more counts
sns.countplot(x="day", data=tips)   # for count: count plot is used -> it counts the number of occurrences of each category in the specified column
```

```python
# we will use join plot 
sns.jointplot(x = tips.total_bill, y = tips.tip) # draw a plot of two variable with distribution > bivariant plot 
# joinplot : understand relationship and also distribution of two variables 
```

```python
# now can we see the above relationship >> with the help of line >> for this we have to use regression (reg)
sns.jointplot(x = tips.total_bill, y = tips.tip, kind = "reg")
plt.show()
```

```python
# using kde :
sns.jointplot(x = tips.total_bill, y = tips.tip, kind = "kde")
# contour plot 
# kde is kernal density estimato 
```

```python
# pair plot : makes plots scatter plot with respect to all numerical columns 
sns.pairplot(tips)
plt.show()
```
