```python
# lets import some dataset from seaborn 
import seaborn as sns
```

```python
# to see all available datasets in seaborn
sns.get_dataset_names()
```

```python
df = sns.load_dataset("tips")
df.head()
```

```python
# covariance and correlation are calculated on numerical columns only 
df.dtypes
```

```python
import pandas as pd
```

```python
df.cov(numeric_only=True)  # covariance matrix
```

```python
# correlation matrix
df.corr(numeric_only= True)
```

```python
# lets plot the correlation matrix using heatmap
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(8,6))
cor = df.corr(numeric_only = True)
sns.heatmap(cor, annot = True, cmap=plt.cm.CMRmap_r)
plt.show()
# annot: annotation -> it shows the values in the boxes 
# cmap: colormap -> it changes the color scheme of the heatmap , plt.cm.CMRmap_r is one of the colormaps available in matplotlib 
```
