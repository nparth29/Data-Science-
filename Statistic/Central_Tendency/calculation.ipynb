```python
import numpy as np
age = np.array([22, 25, 47, 52, 46, 56, 55, 23, 40, 33, 34, 43, 36])
# mean of age 
mean_age = np.mean(age)
print("Mean age:", mean_age)
```

```python
# median of age 
media_age  = np.median(age)
media_age
```

```python
# numpy doesn't have mode function 
# we have to import from scipy
from scipy import stats
stats.mode(age)
```
